Series is similar to array except that it can be accessed via labels

```python
import numpy as np
import pandas as pd
```

```python
labels = ['a','b','c']
my_data = [10,20,30]
arr = np.array(my_data)
d = {'a':10,'b':20,'c':30}
```

```python
pd.Series(data = my_data)
```

```python
pd.Series(data=my_data,index=labels)
```

Unlike numpy array we can modify index names

```python
pd.Series(my_data,labels)
```

```python
pd.Series(arr)
```

```python
pd.Series(arr,labels)
```

```python
d
```

```python
pd.Series(d)
```

Series can hold any data type values or even built in functions like eg below

```python
pd.Series(data=[sum,print,len])
```

Example to grab information from series

```python
ser2 = pd.Series([1,2,5,4],['USA','Germany','Italy','Japan'])
ser2
```
