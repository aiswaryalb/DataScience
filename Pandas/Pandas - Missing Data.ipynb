```python
import numpy as np 
import pandas as pd
```

```python
d = {'A':[1,2,np.nan],'B':[5,np.nan,np.nan],'C':[1,2,3]}
```

```python
df = pd.DataFrame(d)
```

```python
df
```

```python
df.dropna()
#drops any rows with 1 or more NaN value
```

```python
#for dropping columns with null values
df.dropna(axis=1)
```

```python
df.dropna(thresh=2)
#keeps rows with  2 non nan values
```

```python
df
```

```python
#to fill nan values
df.fillna(value='FILL VALUE')
```

```python
df['A'].fillna(value=df['A'].mean())
```
