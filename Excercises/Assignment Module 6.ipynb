For the dataset “Indian_cities”, 
	
Find out top 10 states in female-male sex ratio

```python
df.groupby('state_name').sum()['sex_ratio'].sort_values(ascending = False).head(10)
```

Find out top 10 cities in total number of graduates



```python
df[['name_of_city','total_graduates']].sort_values('total_graduates',ascending = False).head(10)
```

Find out top 10 cities and their locations in respect of  total effective_literacy_rate.


```python
df.sort_values('effective_literacy_rate_total',ascending = False)[['name_of_city','location']].head(10)
```

Construct Boxplot on total effective literacy rate and draw inferences



```python
bp = plt.boxplot(df['effective_literacy_rate_total'],vert = False)
#The median is 85.97
#from box plot we can infer that there are multiple outliers less than minimum which need to be analyzed if they are outliers or not
#from q1,q3, min = q1-1.5IQR = 70.38, values less than this will be outlier
import matplotlib.cbook as cbook
result = cbook.boxplot_stats(df['effective_literacy_rate_total'])
result
```
