Sliced notations

```python
arr[1:5]
```

```python
arr[0:5]
```

```python
arr[2:]
#array from 2 till end
```

```python
arr[:6]
#this is same as arr[0:6]
```

```python
arr[0:5] = 100
arr
#single value can be broadcasted to multiple postions
```

```python
#Restoring arr back to as it was
arr = np.arange(0,11)
arr
```

```python
slice_of_arr = arr[0:6]
slice_of_arr
```

```python
slice_of_arr[:] = 99
slice_of_arr
```

```python
arr
```

Here both slice_of_arr and arr value from 0 to 5 gets changed to 99. This is because it copies memory reference. To get a copy we can do as below. 

```python
arr_copy = arr.copy()
arr_copy
```

```python
arr_copy[:] = 100
arr_copy
```

```python
arr
```

Here we can see that the original array's value is not affected

Indexing 2D array

```python
arr_2d = np.array([[5,10,15],[20,25,30],[35,40,45]])
arr_2d
```

Indexing digit 5

```python
#position will be row * column
#Double bracket notation
arr_2d[0][0]
```

```python
#single bracket notation to index 30
arr_2d[1,2]
```

Grabbing sections of matrix using sliced notation

```python
arr_2d
```

```python
arr_2d[:2,1:]
```

```python
#To retrieve 20 to 45
arr_2d[1:,:]
```

```python
arr1 = np.arange(1,11)
arr1
```

```python
bool_arr = arr1 > 5
bool_arr
```

```python
arr1[bool_arr]
```

```python
#The above statement can be simplified as
arr1[arr1>5]
```

```python
arr1[arr1<3]
```
