```python
import matplotlib.pyplot as plt
```

```python
%matplotlib inline
```

```python
import numpy as np
x = np.linspace(0, 5, 11)
y = x ** 2
```

```python
x
```

```python
y
```

```python
#Functional method
plt.plot(x,y)
plt.xlabel('X Label')
plt.ylabel('Y Label')
plt.title('Title')
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'r')

plt.subplot(1,2,2)
plt.plot(y,x,'b')
```

```python
#Object Oriented
fig = plt.figure()

axes = fig.add_axes([0.1,0.10,0.8,0.8])
axes.plot(x,y)
axes.set_xlabel('X label')
axes.set_ylabel('Y label')
axes.set_title('Title')
```

```python
fig = plt.figure()
axes1 = fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.2,0.5,0.4,0.3])
#left,bottom ,width and height)
axes1.plot(x,y)
axes1.set_title('Larger Plot')

axes2.plot(y,x)
axes2.set_title('Smaller Plot')
```

```python
fig , axes = plt.subplots(nrows = 1 , ncols =2)
#axes.plot(x,y)
axes[0].plot(x,y)
axes[0].set_title('First')

axes[1].plot(y,x)
axes[1].set_title('Second')
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,x**2,label='X Square')
ax.plot(x,x**3,label='X Cube')
ax.legend(loc = 0)
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,color='green')
```

```python
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,color = '#B02BD8',linewidth=3,alpha =0.5,linestyle='-',marker='1',
        markersize=8,markerfacecolor = 'yellow',markeredgewidth=3,markeredgecolor='green')
#alpha for transparency and linewidth can be mentioned as lw also
#for linestyle we can mention as ls also
```

```python
#To set limit of axes
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.plot(x,y,color = '#B02BD8',lw=3,ls='--')
ax.set_xlim([0,1])
ax.set_ylim([0,2])
```
